```python
from dbfread import DBF
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python

# Ruta al archivo DBF
ruta_dbf = '/Users/carolinavega/Documents/UCM/Exclusion/DesercionOficial.dbf'

# Leer el archivo DBF
with DBF(ruta_dbf, encoding='latin-1') as table:
    # Convertir a DataFrame
    oficial = pd.DataFrame(iter(table))
```

```python

# Ruta al archivo DBF
ruta_dbf = '/Users/carolinavega/Documents/Exclusion/TDesercion_NoOficial.dbf'

# Leer el archivo DBF
with DBF(ruta_dbf, encoding='latin-1') as table:
    # Convertir a DataFrame
    nooficial = pd.DataFrame(iter(table))
```

```python
pd.set_option('display.max_columns', None)
```

```python

# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ingresos_1', y='TotalHM', data=oficial, color='green', alpha=0.7, s=100)
plt.title('Relación entre Ingresos y Tasa de Deserción Colegios Oficiales')
plt.xlabel('Ingresos')
plt.ylabel('Tasa de Deserción')
plt.grid(True)
plt.show()
```

```python
# Gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ingresos_1', y='TotalHM', data=nooficial, color='blue', alpha=0.7, s=100)
plt.title('Relación entre Ingresos y Tasa de Deserción de Colegios No Oficial')
plt.xlabel('Ingresos')
plt.ylabel('Tasa de Deserción')
plt.grid(True)
plt.show()
```

```python
# Scatter plot with a reversed semaphore-like color palette
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ingresos_1', y='TotalHM', hue='TotalHM', data=nooficial, alpha=0.7, s=100, palette='RdYlGn_r', edgecolor=None)

plt.title('Relación entre Ingresos y Tasa de Deserción Colegios No Oficiales')
plt.xlabel('Ingresos')
plt.ylabel('Tasa de Deserción')
plt.grid(True)
plt.show()
```
